```python
def euro_option_bs(S,K,T,r,vol,payoff):
    d1=(np.log(S/K)+(r+0.5*vol**2)*T)/(vol*np.sqrt(T))
    d2=(np.log(S/K)+(r-0.5*vol**2)*T)/(vol*np.sqrt(T))
    if payoff=='call':
        option_value=S*si.norm.cdf(d1,0.00,1.00)-K*np.exp(-r*T)*si.norm.cdf(d2,0.00,1.00)
    elif payoff=="put":
        option_value=K*np.exp(-r*T)*cdf(-d2,0.00,1.00)-S*si.norm.cdf(-d1,0.00,1.00)
    return option_value
p=euro_option_bs(100,100,1,0.05,0.25,'call')
print("The option price is", p)
```
